```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def find_pi(n):
    '''Takes in integer n (n points used for integration), does Monte Carlo to find pi using a mask. Returns int pi
    and int error.'''

    # here's Monte Carlo for finding pi
    (x,y) = np.random.uniform(-1,1,size=(2,int(n)))
    ones = np.ones(int(n))
    distance = np.hypot(x,y)
    mask = np.where(distance <= 1.0)
    inside = ones[mask]
    pi = 4*np.sum(inside)/n
    error = np.abs(np.pi - pi)
    
    return pi, error
```

```python
# set up range of points and make an array to collect errors
test = 2**(np.linspace(0, 20, 21)).astype(int)
error2 = np.array([])
# time things
start = time.time()
for i in test:
    pi, error = find_pi(i)
    error2 = np.append(error2, error)
stop = time.time()

print("Time to run:", stop - start, "seconds; pi:", pi)

# plotting things
fig,ax = plt.subplots(1,2)
fig.set_size_inches(8,5)
fig.tight_layout(w_pad=4)

ax[0].set_aspect('equal')
ax[0].set_xlabel("here is x")
ax[0].set_ylabel("here is y")
ax[0].set_title("Here is pi graphically")
ax[0].scatter(my_x, my_y)
ax[0].text(-0.5,-.05, r"$\pi \simeq {0:f}$".format(pi),color='k', fontsize=20)

ax[1].set_xlabel("Number of Steps")
ax[1].set_ylabel("Amount of Error")
ax[1].set_title("Error vs steps");
ax[1].scatter(np.log(test), np.log(error2), label="Error")

y = test**(-(1.0/2.0))
    
ax[1].plot(np.log(test),np.log(y), color='red', label="MC error trend");
ax[1].legend(loc=0);
```
